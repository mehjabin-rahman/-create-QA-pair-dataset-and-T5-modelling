```python
from transformers import pipeline

qa_model = pipeline("question-answering")

question = "Who is Crones Mill and what is their role in the story?"
context = "Red Fork he reduced Bowshot Bridge to embers burned Old Ferry and Crones Mill destroyed the motherhouse at Bechester always vanishing back into the sky before the hunters could arrive Vhagar never lingered nor did the survivors oft agree on which way the dragon had flown"
qa_model(question = question, context = context)
## {'answer': 'behind the chair', 'end': 61, 'score': 0.8110504746437073, 
#   'start': 45}
```

```python
from transformers import pipeline

qa_model = pipeline("question-answering")
def extract_answer(row):
    result = qa_model(question=row['questions'], context=row['context'])
    row['answer'] = result['answer']
    row['start'] = result['start']
    row['end'] = result['end']
    row['score'] = result['score']
    return row

# Apply the function to each row of the dataframe
df = df.apply(extract_answer, axis=1)

# Save the updated dataframe to a new csv file
df.to_csv('new_answer_dataset.csv', index=False)
```
